## Data Analysis and Visualization

#### Gender Distribution

```python
gender = project_pa['Gender'].value_counts()
```

```python
palette_color = sns.color_palette('bright')

labels = ['Female', 'Male']
plt.pie(gender, labels=labels, autopct='%.0f%%')
plt.title('Gender distribution of Students')
plt.show()
```

```python
female_df = project_pa[project_pa.Gender=='female']
male_df = project_pa[project_pa.Gender=='male']
female_df.columns
```

#### Age distribution in students

```python
female_age = female_df['Age'].value_counts()
plt.figure(figsize=(7, 6))
ax = female_age.plot(kind='bar', rot=0, color='c')
ax.set_title('Age Distribution in female students')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width()
    space = 1
    label = format(y_value)
    ax.annotate(label, (x_value, y_value))
    
plt.show()
```

```python
male_age = male_df['Age'].value_counts()
plt.figure(figsize=(7, 6))
ax = male_age.plot(kind='bar', rot=0, color='c')
ax.set_title('Age Distribution in male students')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width()
    space = 1
    label = format(y_value)
    ax.annotate(label, (x_value, y_value))
    
plt.show()
```
