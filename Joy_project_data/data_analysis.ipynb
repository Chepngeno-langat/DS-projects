#### Percentiles

```python
bmi_percentiles = np.percentile(data_df['BMI'], [5, 85, 95])
```

```python
def categorize_bmi(bmi):
    if bmi < bmi_percentiles[0]:
        return 'Below 5th percentile'
    elif bmi_percentiles[0] <= bmi < bmi_percentiles[1]:
        return '5th to 85th percentile'
    elif bmi_percentiles[1] <= bmi < bmi_percentiles[2]:
        return '86th to 95th percentile'
    else:
        return 'Above 95th percentile'

# Apply categorization to create a new column
data_df['BMI Percentile'] = data_df['BMI'].apply(categorize_bmi)
data_df
```

```python
data_df['BMI Percentile'].value_counts()
```

```python
BMI_percentiles = data_df['BMI Percentile'].value_counts()
labels = ['5th to 85th percentile', '86th to 95th percentile', 'Above 95th percentile', 'Below 5th percentile']
plt.pie(BMI_percentiles, labels=labels, autopct='%.0f%%')
plt.title('BMI Percentiles')
plt.show()
```
