```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
train_data['VIP'].value_counts()
```

```python
train_data['VIP'].fillna(method='ffill', axis=0, inplace=True)
train_data['VIP'].value_counts()
```

```python
train_data['RoomService'] = train_data['RoomService'].fillna(train_data['RoomService'].mode()[0])
train_data['FoodCourt'] = train_data['FoodCourt'].fillna(train_data['FoodCourt'].mode()[0])
train_data['ShoppingMall'] = train_data['ShoppingMall'].fillna(train_data['ShoppingMall'].mode()[0])
train_data['Spa'] = train_data['Spa'].fillna(train_data['Spa'].mode()[0])
train_data['VRDeck'] = train_data['VRDeck'].fillna(train_data['VRDeck'].mode()[0])
```

```python
train_data['Age'] = train_data['Age'].fillna(np.mean(train_data['Age']))
```

```python
train_data.isnull().sum()
```

```python
train_data = train_data.drop('Cabin', axis=1)
train_data = train_data.drop('Destination', axis=1)
train_data = train_data.drop('Name', axis=1)
```

```python
train_data
```

```python
train_data['Transported'].value_counts()
```

```python
#train_data['HomePlanet'] = train_data['HomePlanet'].apply(lambda x: 1 if x == 'Earth' else 0)
train_data['CryoSleep'] = train_data['CryoSleep'].replace({True: 1, False: 0})
train_data['VIP'] = train_data['VIP'].replace({True: 1, False: 0})
train_data['Transported'] = train_data['Transported'].replace({True: 1, False: 0})
```

```python
train_data
```

```python
dummies = pd.get_dummies(train_data['HomePlanet'])
dummies
```

```python
train_data = pd.concat([train_data,dummies],axis=1)
```

```python
train_data = train_data.drop(['HomePlanet'], axis=1)
#train_data = train_data.drop(['PassengerId'], axis=1)
train_data
```

```python
train_data = train_data.drop(['Mars'], axis=1)
```

```python
# import seaborn as sns
# sns.histplot()
```

```python
X = train_data.drop('Transported', axis=1)
y = train_data['Transported']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
train_model = DecisionTreeClassifier(max_depth=9)
train_model.fit(X_train, y_train)

# from sklearn.ensemble import RandomForestClassifier
# train_model = RandomForestClassifier()
# train_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print('Training accuracy...', accuracy_score(y_train, train_model.predict(X_train)))
print('Test accuracy...', accuracy_score(y_test, train_model.predict(X_test)))
```
