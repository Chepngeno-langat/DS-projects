```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
vehicle_data = pd.read_csv('data.csv')
vehicle_data.head()
```

```python
vehicle_data.tail()
```

```python
vehicle_data.dtypes
```

```python
vehicle_data.info()
```

```python
vehicle_data = vehicle_data.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
vehicle_data
```

```python
vehicle_data = vehicle_data.rename(columns={'Engine HP': 'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission', 'Driven_Wheels': 'Drive Mode','highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price' })
vehicle_data
```

```python
#check and remove duplicates
duplicate_rows_data = vehicle_data[vehicle_data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_data.shape)
```

```python
vehicle_data = vehicle_data.drop_duplicates()
vehicle_data.info()
```

```python
#check for missing data
vehicle_data.isnull().sum()
```

```python
vehicle_data = vehicle_data.dropna()
vehicle_data.isnull().sum()
```

```python
#checking for outliers
sns.boxplot(y=vehicle_data['Price'])
```

```python
sns.boxplot(y=vehicle_data['HP'])
```

```python
sns.boxplot(y=vehicle_data['Cylinders'])
```

```python
Q1 = vehicle_data.quantile(0.25)
Q3 = vehicle_data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
vehicle_data = vehicle_data[~((vehicle_data < (Q1-1.5 * IQR)) |(vehicle_data > (Q3 + 1.5 * IQR))).any(axis=1)]
vehicle_data.shape
```

```python
vehicle_data.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of cars by make')
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(20,10))
c= vehicle_data.corr()
sns.heatmap(c,cmap='BrBG',annot=True)
c
```

```python
#scatterplot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(vehicle_data['HP'], vehicle_data['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
